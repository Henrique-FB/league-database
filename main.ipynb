```python
from riotwatcher import LolWatcher, ApiError
from configparser import ConfigParser
```

```python
#Setting up confiigurations
config = ConfigParser()
config.read('config.ini')

RIOT_API_KEY = config.get('apikeys','riot')
```

```python


lol_watcher = LolWatcher(RIOT_API_KEY)

my_region = 'br1'
 


player = lol_watcher.summoner.by_name(my_region, 'taxi drive')


print(player)

```

```python
matches = lol_watcher.match.matchlist_by_puuid(my_region,player['puuid'])
print(matches)
```

```python
test = lol_watcher.match.timeline_by_match(my_region, matches[1])
```
